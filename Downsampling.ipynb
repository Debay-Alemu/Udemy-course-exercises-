# Downsampling



Downsampling


```python
import torch
import torch.nn.functional as F
```

```python
image = torch.tensor([
    [1,3,2,4],
    [5,6,7,8],
    [9,2,3,1],
    [4,0,6,2]]).float().unsqueeze(0).unsqueeze(0)
```

```python
image.ndim
```

```python
image.shape
```

```python
# Max pooling operation
pooled = F.max_pool2d(image, kernel_size=2)
```

```python
conv_layer = torch.nn.Conv2d(1, 1, kernel_size=2, stride=1)
conv_output = conv_layer(image)
```

```python
print("Max-Pooled Output : ", pooled.squeeze())
```

```python
print("Strided Convolution Output : ", conv_output.squeeze())
```
