Upsampling

```python
import torch
import torch.nn.functional as F
```

```python
image = torch.tensor([
    [1,2],
    [3,4]]).float().unsqueeze(0).unsqueeze(0)
```

```python
# Nearest-neighbour upsampling
upsampled_nn = F.interpolate(image, scale_factor=2, mode='nearest')
```

```python
# Transposed Convolution
conv_transpose = torch.nn.ConvTranspose2d(1, 1, kernel_size=3, stride=2, padding=1)
output_transpose = conv_transpose(image)
```

```python
print("Nearest-neighbour upsampling :", upsampled_nn.squeeze())
print("Transpose Convolution output :", upsampled_nn.squeeze())
```
